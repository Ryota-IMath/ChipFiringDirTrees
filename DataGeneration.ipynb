```python
from itertools import product
# Helper from stackexchange. Computes Dyck Paths
def solve(num, cache={0: ['']}):
    if num not in cache:
        cache[num] = ['(%s)%s' % t for i in range(1, num + 1)
            for t in product(solve(i - 1), solve(num - i))]
    return cache[num]
def PossibleFinalPositions(n, m, prevDict={}):
    memoizedVersion = False
    if prevDict!={}:
        print("Using memoization")
        memoizedVersion = True
        
    if m >= 2 ** n or m < 0:
        return set()
    else:
        if m == 2 ** n -1:
            return set([2 ** n])
        elif m == 0:
            return set([1])
        else:
            paths = solve(2 ** (n-1))
            s = set()
            for p in paths:
                if p[m] == '(':
                    mprime = 0
                    counter = 0
                    while counter < (2 ** n) and counter < m:
                        if p[counter] == '(':
                            mprime +=1
                        counter += 1
                    if prevDict == {}:
                        s = s.union(PossibleFinalPositions(n-1, mprime))
                    else:
                        s = s.union(prevDict[mprime])
                else:
                    mprime = 0
                    counter = 0
                    while counter < (2 ** n) and counter < m:
                        if p[counter] == ')':
                            mprime +=1
                        counter += 1
                    if prevDict  == {}:
                        sp = PossibleFinalPositions(n-1, mprime)
                    else:
                        sp = prevDict[mprime]
                    sprime = [t+2 ** (n-1) for t in sp]
                    s = s.union(set(sprime))
            return s
```

```python
# Do the more direct method
def solve3(num, collection_of_paths):
    if num == 0:
        return ""
    if num == 1:
        return "abc"
    else:
        resulting = set()
        for i in range(1, 3*num):
            for p in collection_of_paths:
                resulting.add("a"+p[:i-1]+"b"+p[i-1:]+"c")
                resulting.add("abc"+p)
        return resulting
```

```python
from itertools import permutations
def solve3(m):
    s = ""
    for i in range(m):
        s = s+ "abc"
    perms = set([''.join(p) for p in permutations(s)])
    return perms

def PossibleFinalPositions3(n, m, prevDict={}):
    memoizedVersion = False
    if prevDict!={}:
        print("Using memoization")
        memoizedVersion = True
        
    if m >= 3 ** n or m < 0:
        return set()
    else:
        if m == 3 ** n -1:
            return set([3 ** n])
        elif m == 0:
            return set([1])
        else:
            paths = solve(3 ** (n-1))
            s = set()
            for p in paths:
                if p[m] == 'a':
                    mprime = 0
                    counter = 0
                    while counter < (3 ** n) and counter < m:
                        if p[counter] == 'a':
                            mprime +=1
                        counter += 1
                    if prevDict == {}:
                        s = s.union(PossibleFinalPositions3(n-1, mprime))
                    else:
                        s = s.union(prevDict[mprime])
                elif p[m] == 'b':
                    mprime = 0
                    counter = 0
                    while counter < (3 ** n) and counter < m:
                        if p[counter] == 'b':
                            mprime +=1
                        counter += 1
                    if prevDict == {}:
                        sp = (PossibleFinalPositions3(n-1, mprime))
                    else:
                        sp = (prevDict[mprime])
                    sprime = [t+3 ** (n-1) for t in sp]
                    s = s.union(set(sprime))
                else:
                    mprime = 0
                    counter = 0
                    while counter < (3 ** n) and counter < m:
                        if p[counter] == 'c':
                            mprime +=1
                        counter += 1
                    if prevDict == {}:
                        sp = (PossibleFinalPositions3(n-1, mprime))
                    else:
                        sp = (prevDict[mprime])
                    sprime = [t+2*(3 ** (n-1)) for t in sp]
                    s = s.union(set(sprime))
            return s
```
